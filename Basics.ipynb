# Star Wars Survey

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Storing the csv to a datafram

```python
star = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

```python
star.head()
```

```python
star.columns
```

#### Removing columns where respondentid is null

```python
star = star[star['RespondentID'].notnull()]
```

```python
star.head(10)
```

```python
star['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts(dropna=False)
```

```python
star['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts(dropna=False)
```

### Converting the yes/no columns to boolean

```python
mapping = {
    'Yes': True,
    'No': False
}
```

```python
star['Have you seen any of the 6 films in the Star Wars franchise?'] = star['Have you seen any of the 6 films in the Star Wars franchise?'].map(mapping)
```

```python
star['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(mapping)
```

### Converting the question regarding each film to True or False boolean mask

```python
print(star.iloc[0])
```

```python
movie_map = {
    'Star Wars: Episode I  The Phantom Menace': True,
    'Star Wars: Episode II  Attack of the Clones': True,
    'Star Wars: Episode III  Revenge of the Sith': True,
    'Star Wars: Episode IV  A New Hope': True,
    'Star Wars: Episode V The Empire Strikes Back': True,
    'Star Wars: Episode VI Return of the Jedi': True,
    np.NaN: False
}
```

```python
for col in star.columns[3:9]:
    star[col] = star[col].map(movie_map)
```

```python
star.head(10)
```

### Renaming the columns 3 to 9 to something meaningful

```python
star.columns
```

```python
name_map = {
    'Which of the following Star Wars films have you seen? Please select all that apply.': 'seen_1',
    'Unnamed: 4': 'seen_2',
    'Unnamed: 5': 'seen_3',
    'Unnamed: 6': 'seen_4',
    'Unnamed: 7': 'seen_5',
    'Unnamed: 8': 'seen_6'
}
```

```python
star = star.rename(name_map,axis=1)
```

```python
star.columns
```

### Converting the columns 9 to 15 to numeric datatype

```python
star[star.columns[9:15]] = star[star.columns[9:15]].astype(float)
```

```python
next_name_map = {
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': 'ranking_1',
    'Unnamed: 10': 'ranking_2',
    'Unnamed: 11': 'ranking_3',
    'Unnamed: 12': 'ranking_4',
    'Unnamed: 13': 'ranking_5',
    'Unnamed: 14': 'ranking_6'
}
```

```python
star = star.rename(next_name_map,axis=1)
```

```python
star.columns
```

### Calculating mean of each movie ranking column

```python
rank = star[star.columns[9:15]].mean()
```

```python
fig, ax = plt.subplots(figsize=(10,3))
ax.bar(range(6),rank)
ax.set_xticklabels(['','ranking_1','ranking_2','ranking_3','ranking_4','ranking_5','ranking_6'])
plt.show()
```

#### So according to the survey, ranking_5 is supposed to be the best movie

## Calculating the number of ratings for each movie

```python
number = star[star.columns[3:9]].sum()
```

```python
fig, ax = plt.subplots(figsize=(10,3))
ax.bar(range(6),number)
ax.set_xticklabels(['','seen_1','seen_2','seen_3','seen_4','seen_5','seen_6'])
plt.show()
```

### It seems that the movies which are rated the most have got a better rating as well

## Grouping various columns of the dataframe

```python
star['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts(dropna=False)
```

```python
star_fan = star[star['Do you consider yourself to be a fan of the Star Wars film franchise?']==True]
star_fan_not = star[star['Do you consider yourself to be a fan of the Star Wars film franchise?']==False]
```

```python
new_rank1 = star_fan[star_fan.columns[9:15]].mean()
new_rank2 = star_fan_not[star_fan_not.columns[9:15]].mean()
```

```python
fig, ax = plt.subplots(figsize=(10,3))
ax.bar(range(6),new_rank1)
ax.set_xticklabels(['','ranking_1','ranking_2','ranking_3','ranking_4','ranking_5','ranking_6'])
ax.set_xlabel('rank_by_fan')
plt.show()

fig, ax = plt.subplots(figsize=(10,3))
ax.bar(range(6),new_rank2)
ax.set_xlabel('rank_by_non_fan')
ax.set_xticklabels(['','ranking_1','ranking_2','ranking_3','ranking_4','ranking_5','ranking_6'])
plt.show()
```

```python
new_num1 = star_fan[star_fan.columns[3:9]].mean()
new_num2 = star_fan_not[star_fan_not.columns[3:9]].mean()
```

```python
fig, ax = plt.subplots(figsize=(10,3))
ax.bar(range(6),new_num1)
ax.set_xticklabels(['','seen_1','seen_2','seen_3','seen_4','seen_5','seen_6'])
ax.set_xlabel('seen_by_fan')
plt.show()

fig, ax = plt.subplots(figsize=(10,3))
ax.bar(range(6),new_num2)
ax.set_xlabel('seen_by_non_fan')
ax.set_xticklabels(['','seen_1','seen_2','seen_3','seen_4','seen_5','seen_6'])
plt.show()
```
